```python
from langchain.embeddings import HuggingFaceInstructEmbeddings
from langchain.llms import HuggingFaceHub
from langchain.vectorstores import Chroma
from langchain.embeddings import OpenAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain.document_loaders import TextLoader
from langchain.document_loaders import DirectoryLoader
```

## Load multiple and process documents

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from langchain.document_loaders.pdf import PyPDFDirectoryLoader
# Load and process the text files
#loader = DirectoryLoader("google drive path or local folder path", glob = "./*.pdf, loader_cls = pypdf")
# loader = TextLoader('single_text_file.txt')
loader = PyPDFDirectoryLoader('/content/drive/MyDrive/UoM', glob="./*.pdf")
#loader = DirectoryLoader('/content/drive/MyDrive/UoM', glob="./*.pdf", loader_cls= Py)

documents = loader.load()
```

```python
#splitting the text into samller chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
texts = text_splitter.split_documents(documents)
```

```python
len(texts)
```

```python
texts[3]
```

## create the DB

```python
#Create open source embeddings
embeddings = HuggingFaceInstructEmbeddings(model_name="hkunlp/instructor-xl")
```

```python
# Embed and store the texts
# Supplying a persist_directory will store the embeddings on disk
persist_directory = 'db'

#OpenAI embeddings
#embedding = OpenAIEmbeddings()

vectordb = Chroma.from_documents(documents=texts,
                                 embedding=embeddings,
                                 persist_directory=persist_directory)
```

```python
# persiste the db to disk
vectordb.persist()
vectordb = None
```

```python
#load the persisted database from disk, and use it as normal.
vectordb = Chroma(persist_directory=persist_directory,
                  embedding_function=embeddings)
```

## Make a retriever

```python
retriever = vectordb.as_retriever()
```

```python
docs = retriever.get_relevant_documents("when was university of miskolc established?")
```

## Make a chain

```python
# create the chain to answer questions
#for OpenAI assign llm = OpenAI()
#Local llm, huggingfacehub_api_token = "your hugging face access token"
llm = HuggingFaceHub(repo_id="tiiuae/falcon-7b",huggingfacehub_api_token='*****************', model_kwargs={"temperature":0.5, "max_length":512})
qa_chain = RetrievalQA.from_chain_type(llm= llm,
                                  chain_type="map_reduce",
                                  retriever=retriever,
                                  return_source_documents=True)
```

```python
## Cite sources
def process_llm_response(llm_response):
    print(llm_response['result'])
    print('\n\nSources:')
    for source in llm_response["source_documents"]:
        print(source.metadata['source'])
```

```python
# full example
query = "when was the university of miskolc established?"
llm_response = qa_chain(query)
process_llm_response(llm_response)
```
